# Allows to fetch date and time stamp for instagram post 

```python
import instaloader
from datetime import datetime
import pandas as pd
import os
```

## Login session

```python
# Initialize Instaloader instance
L = instaloader.Instaloader()

# Log in to Instagram (if needed, optional)
L.login('#', '#')  # Replace with your Instagram username and password
```

###Mention the page 

```python

# Define the Instagram profile to scrape (e.g., "nike")
profile_name = "#"
```

```python

# Load the profile
profile = instaloader.Profile.from_username(L.context, profile_name)

# Define the date range (October 1 to October 24, 2024)
start_date = datetime(2024, 6, 1)
end_date = datetime(2024, 10, 24)

# Initialize lists to store the filtered post data
post_dates = []
post_times = []

# Loop over the profile's posts
for post in profile.get_posts():
    # Get the post's date
    post_date = post.date
    print(f"Post date: {post_date}")  # Debugging: Print each post's date

    # Stop the loop if we encounter a post before the start date
    if post_date < start_date:
        print("Reached posts earlier than the start date, stopping.")
        break

    # Filter posts within the date range
    if start_date <= post_date <= end_date:
        print(f"Post within range: {post_date}")  # Debugging: Print posts within the date range

        # Append the post's date and time
        post_dates.append(post_date.strftime('%Y-%m-%d'))
        post_times.append(post_date.strftime('%H:%M:%S'))

# Check if any posts were found in the date range
if post_dates:
    # Create a pandas DataFrame with the scraped data
    df = pd.DataFrame({
        'Date': post_dates,
        'Time': post_times
    })

    # Sort the DataFrame by Date in reverse chronological order (latest to earliest)
    df['Date'] = pd.to_datetime(df['Date'])  # Ensure 'Date' is in datetime format for proper sorting
    df = df.sort_values(by='Date', ascending=False)

    # Check the current working directory
    current_dir = os.getcwd()
    print(f"Saving the file in the current directory: {current_dir}")

    # Save the DataFrame to a CSV file
    output_file = f'{profile_name}.csv'
    df.to_csv(output_file, index=False)

    print(f"Data has been saved to {output_file}")
else:
    print("No posts found in the specified date range.")
profile_name = "your_profile_name"
```
